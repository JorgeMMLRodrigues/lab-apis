```python
results = sp.search(q='Los Fabulosos Cadillacs', limit=50)

# Extract popularity of each track
popularities = [track['popularity'] for track in results['tracks']['items']]

print(popularities)
```

```python
# Your answer here
artists = ["Los Fabulosos Cadillacs", "Manu Chao", "Muchachito Bombo Infierno"]


"""
 **Fetch Top Tracks**:
    - Write a function named `get_top_tracks`.
    - This function should accept an artist's name and return the name of the first 5 top tracks by that artist.
    - Use the function `get_top_tracks` to get the first 5 top tracks for each artist in your `artists` list and store the results in a new list named `top_tracks_list`.
"""

def get_top_tracks(artists):

    top_tracks_list = []
    
    for artist in artists:
        results = sp.search(q=artist, limit=50)
        
    
        tracks = [
            (track['name'], track['popularity']) 
            for track in results['tracks']['items']
        ]
        
        top_tracks = [track[0] for track in sorted(tracks, key=lambda x: x[1], reverse=True)[0:5]]
        
        top_tracks_list.append({artist: top_tracks})
    
    return top_tracks_list
    

get_top_tracks(artists)
```

```python
def find_similar_artists(artist_name):
    try:
        results = sp.search(q=artist_name, type='artist', limit=1)
     
        artist = results['artists']['items'][0]

        artist_id = artist['id']
        
        genres = artist.get('genres', [])

        genre_query = genres[0]  

        top_artists = sp.search(q=f"genre:{genre_query}", type='artist', limit=5)

        similar_artist_names = [artist['name'] for artist in top_artists['artists']['items']]

        return {artist_name: similar_artist_names}

    except spotipy.SpotifyException as e:

        print(f"Spotify API Error: {e}")

        return {artist_name: []}

print(find_similar_artists("Jack Johnson"))

```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Exploration**: 
   - Fetch the list of Spotify' 50 most recent tracks. 
   - Extract and display the artist names and their track names.
   - Create a dictionary that shows artist name as keys and inside it a new dictionary that contains the track name and as a value the track name.
   
2. **Deep Dive**:
   - From the resulting dict fetch the tracks popularity
   - Add the popularity key and it corresponding value from to the previous dict (or create a copy from the previous dict)
   - Rank the tracks in descending order of popularity.

4. **Track-tastic**:
   - Extract and display the names of the first 10 tracks


   

```python
# Fetch new album releases
results = sp.new_releases(limit=50)

# Initialize a list to store track details
track_details = []

# Iterate through each album
for album in results['albums']['items']:
    # Fetch tracks for the album
    album_tracks = sp.album_tracks(album['id'])
    for track in album_tracks['items']:
        track_info = {
            'name': track['name'],
            'artists': ', '.join(artist['name'] for artist in track['artists']),
            'album': album['name'],
            'release_date': album['release_date'],
            'external_url': track['external_urls']['spotify']
        }
        track_details.append(track_info)

```

```python
track_details
```

```python
artist_track_dict = {}

for track in track_details:
    artists = track['artists'].split(',')  # Handle multiple artists
    for artist in artists:
        artist = artist.strip()  # Clean any extra spaces around artist names
        artist_track_dict[artist] = track['name']
```

```python
artist_track_dict
```

```python
for artist, track_name in artist_track_dict.items():
    try:
        # Search for the track
        results = sp.search(q=track_name, type='track', limit=1)

        # Get the track details
        if results['tracks']['items']:
            track = results['tracks']['items'][0]
            track_popularity = track['popularity']

            # Update the dictionary with track name and popularity
            artist_track_dict[artist] = {'track_name': track_name, 'popularity': track_popularity}
        else:
            artist_track_dict[artist] = {'track_name': track_name, 'popularity': 'Not Found'}
    except Exception as e:
        print(f"Error processing {track_name}: {e}")
        artist_track_dict[artist] = {'track_name': track_name, 'popularity': 'Error'}
```

```python
artist_track_dict
```

```python
ranked_artist_track_dict = artist_track_dict.copy()
```

```python
ranked_artist_track_dict ={k: v for k, v in sorted(ranked_artist_track_dict.items(), key=lambda item: item[1]['popularity'], reverse=True)}
```

```python
first_10_tracks = [v['track_name'] for k, v in list(ranked_artist_track_dict.items())[:10]]

print(first_10_tracks)
```
